```python
multi_df = df.groupby(['region', 'product']).agg({
    'sales': 'sum',
    'price': 'mean'
})

multi_df
```

```python
#Переставить местами уровни индекса.
swapped_df = multi_df.swaplevel('region', 'product')
swapped_df.sort_index()
swapped_df
```

```python
#Транспонировать таблицу (или создать новую другую) с MultiIndex.
transposed_df = multi_df.T
transposed_df
```

```python
#Продемонстировать работу .merge.
np.random.seed(42)

n_orders = 100
order_ids = range(1001, 1001 + n_orders)
product_ids = np.random.randint(1, 6, size=n_orders)
quantities = np.random.randint(1, 20, size=n_orders)
dates = pd.date_range(start='2023-01-01', periods=n_orders, freq='D')


sales_df = pd.DataFrame({
    'order_id': order_ids,
    'product_id': product_ids,
    'quantity': quantities,
    'date': dates
})

sales_df
```

```python
n_products = 6
product_ids = range(1, n_products + 1)
product_names = ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard', 'Mouse']
prices = np.round(np.random.uniform(50, 1000, size=n_products), 2)
categories = np.random.choice(['Electronics', 'Accessories'], size=n_products)


products_df = pd.DataFrame({
    'product_id': product_ids,
    'product_name': product_names,
    'price': prices,
    'category': categories
})


products_df
```

```python
inner_merge = sales_df.merge(products_df, on='product_id', how='inner')
print("\nInner Merge:")
inner_merge
```

```python
outer_merge = sales_df.merge(products_df, on='product_id', how='outer')
print("\nOuter Merge:")
outer_merge
```

```python
left_merge = sales_df.merge(products_df, on='product_id', how='left')
print("\nLeft Merge:")
left_merge
```

```python
right_merge = sales_df.merge(products_df, on='product_id', how='right')
print("\nRight Merge:")
right_merge
```
